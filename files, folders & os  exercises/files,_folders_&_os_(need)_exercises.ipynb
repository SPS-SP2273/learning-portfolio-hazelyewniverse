```python
import os
import glob
import shutil
```

# Task 1

```python
# Given dictionary for month conversion
months_dict = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
}

# Sample filename to convert
filename = "oct-08_singapore.txt"

# Conversion function
def convert_filename(filename):
    parts = filename.split('-')
    month, day_city = parts[0], parts[1]
    day, city = day_city.split('_')
    new_filename = f"{months_dict[month]}-{day}-{city}"
    return new_filename

# Convert the sample filename
converted_filename = convert_filename(filename)
converted_filename
```

# Task 2

```python
# Function to rename a single file based on the provided old file name
def rename_my_file(old_file_name):
    # Use the existing conversion function
    new_file_name = convert_filename(old_file_name)
    return new_file_name

# Test the function with a sample filename
test_filename = "dec-25_tokyo.txt"
renamed_file = rename_my_file(test_filename)
renamed_file
```

```python
# List of file names to convert
file_list = [
    'oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
    'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
    'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
    'jan-01_tokyo.txt'
]

# Apply the rename_my_file function to each file name in the list
for old_file_name in file_list:
    new_file_name = rename_my_file(old_file_name)
    print(f"{old_file_name} ----> {new_file_name}")
```

# Task 4

```python
# Modified rename_my_file function
def rename_my_file(old_file_path):
    file_path, file_name = old_file_path.split(os.path.sep)
    # Check if file name already changed 
    if "_" not in file_name: 
        new_file_name = file_name 
    else: 
        new_file_name = convert_filename(file_name)
    return f"{file_path}/{new_file_name}"
```

```python
all_file_paths = glob.glob('**/os-collaboration-exercise-data/*.txt', recursive=True)
for file_path in all_file_paths: 
    os.rename(file_path, rename_my_file(file_path))
```

# Task 6

```python
base_dir = 'os-collaboration-exercise-data_sortedbycities'

# Specify the city you're interested in
city = 'Cairo'

# Create a directory for the city if it doesn't already exist
city_dir = os.path.join(base_dir, city)

# Define the directory where the original files are located
source_dir = '/os-collaboration-exercise-data'

# Use glob to find all files for the specified city and copy them to the city's directory
for file_path in glob.glob(f"**{source_dir}/*{city.lower()}*.txt", recursive = True):
    new_file_path = os.path.join(base_dir, city)
    shutil.copy(file_path, new_file_path)

# Print completion message
print(f"Files related to {city} have been copied to {city_dir}.")
```
